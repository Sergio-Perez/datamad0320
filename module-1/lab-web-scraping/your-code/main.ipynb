####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url_Earth = 'https://www.emsc-csem.org/Earthquake/'
````

````python
res_earth= requests.get(url_Earth)
soup_earth = BeautifulSoup(res_earth.text)
earth = soup_earth.select("#tbody tr td.tabev1")
earth[0].text
````

````python
earth2 = soup_earth.select("#tbody tr")
earth2[0].text.split()
````

````python
earth3= soup_earth.find("tbody").find_all("tr")[1:]
print(earth3[0])
````

````python
def terremotos(tag):
    td6 = tag.select("td.tabev6")
    td1 = tag.select("td.tabev1")
    td2 = tag.select("td.tabev2")
    tdr = tag.select("td.tb_region")
    return { 
        "date" : td6[0].find("a").text.split()[0],
        "time" : td6[0].find("a").text.split()[1],
        "latitude" : td1[0].text + td2[0].text,
        "longitude" : td1[1].text + td2[1].text,
        "region" :  tdr[0].text
    }
terremotos(earth3[0])
````

````python
terremotos_earth = list(map(terremotos,earth3))
pd =pd.DataFrame(terremotos_earth) 
display(pd)
````

#### Count number of tweets by a given Twitter account.

You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
#your code
cuenta = input('Cuenta sobre la que quieres contar los tweets: ')
res = requests.get(url+cuenta)
soup = BeautifulSoup(res.text)

try:
    tweet_info = soup.find("li", {"class":"ProfileNav-item ProfileNav-item--tweets is-active"})
    tweets = tweet_info.find("a").find("span", {"class":"ProfileNav-value"})
    print("{} tiene {} tweets.".format(cuenta,tweets.get("data-count")))

except:
    print('Account name not found...')
````
